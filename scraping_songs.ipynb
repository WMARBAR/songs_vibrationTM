# DATABASE CREATOR

```python
import pandas as pd
# Lista de todos tus DataFrames
dataframes = [df_shakira, df_baladas, df_reggeton, df_ranchera,df_merengue,df_maluma,df_arcangel,df_don_omar,df_romeo_santos]

# Crear la lista total de columnas
# Primero las columnas fijas
columnas_fijas = ['song_name', 'author_name', 'Genero']

# Luego todas las columnas de estrofas hasta la cantidad máxima detectada
# Detectar el máximo número de estrofa en todas las bases
max_estofras = 0
for df in dataframes:
    columnas_estrofas = [col for col in df.columns if col.startswith('estrofa')]
    max_estofras = max(max_estofras, len(columnas_estrofas))

# Generar la lista completa de columnas que tendrá la base final
columnas_finales = columnas_fijas + [f'estrofa{i}' for i in range(1, max_estofras + 1)]

# Ahora aseguramos que cada DataFrame tenga todas esas columnas
dataframes_formateados = []
for df in dataframes:
    # Crear las columnas faltantes como NaN
    for col in columnas_finales:
        if col not in df.columns:
            df[col] = pd.NA
    # Reordenar columnas
    df = df[columnas_finales]
    dataframes_formateados.append(df)

# Concatenar todo en una sola base
df_total = pd.concat(dataframes_formateados, ignore_index=True)



print('Base general creada exitosamente. Columnas totales:', len(df_total.columns))
```

# LIMPIEZA DE DATA

```python
import re

def limpiar_para_sentimiento(texto):
    if pd.isnull(texto):
        return texto
    # Reemplazar saltos de línea por espacio
    texto = texto.replace('\n', ' ')
    # Eliminar múltiples espacios
    texto = re.sub(r'\s+', ' ', texto)
    # Eliminar caracteres muy raros (opcional)
    texto = re.sub(r'[^\w\sáéíóúÁÉÍÓÚñÑ.,!?¡¿]', '', texto)
    return texto.strip()

for col in df_total.columns:
    if col.startswith('estrofa'):
        df_total[col] = df_total[col].apply(limpiar_para_sentimiento)

```

# SENTIMIENTO

```python
from transformers import pipeline

def analizar_sentimiento_bert(df, columna_texto):
    classifier = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

    sentimientos = []
    scores = []

    for texto in df[columna_texto]:
        if pd.isnull(texto):
            sentimientos.append(None)
            scores.append(None)
        else:
            resultado = classifier(texto)[0]
            sentimientos.append(resultado['label'])  # Ej: '1 star', '5 stars'
            scores.append(resultado['score'])        # Confianza del modelo

    columna_sent = f"sentimiento_{columna_texto}"
    columna_score = f"score_{columna_texto}"

    df[columna_sent] = sentimientos
    df[columna_score] = scores

    return df
```

```python
df_total = analizar_sentimiento_bert(df_total, 'estrofa4')
```
